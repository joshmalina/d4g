```python
# want to look longitudinally at districts that seem to be 1) closing the gap
# so schools where the difference in average score by race is decreasing over time
read_gap = pd.read_csv('/Users/jm/Downloads/Data4Good_Samples/AchievementGap_Reading.csv')
read_gap.head()
```

```python
# grab the StudentGroup field of the 4th row
read_gap.loc[4, 'StudentGroup']
# interprets indices as labels, not positions, so literally the row named "4", the row named "5", etc ...
read_gap.loc[4:10, 'DistrictName']
```

```python
# unlike loc, iloc is not by row name or column name, but by row position and column position
# doesn't grab the item with position 10, just those in position 4, 5, 6, 7, 8, 9
read_gap.iloc[4:10, 5]
```

```python
# show mean reading deficits by ethnic group
read_gap_sub = read_gap[['StudentGroup', 'ProficiencyIndex']]
grouped = read_gap_sub.groupby('StudentGroup')
group_means = grouped['ProficiencyIndex'].mean().sort_values()
group_means_df = pd.DataFrame([group_means.index, group_means.values]).T
group_means_df.columns = ['StudentGroup', 'MeanReadProficiency']
group_means_df
```

```python
ggplot(group_means_df, aes(x='StudentGroup', weight='MeanReadProficiency', fill='StudentGroup')) + \
    geom_bar() + \
    scale_y_continuous(labels='comma') + \
    ggtitle('Mean Reading Proficiency by Student Group')
```
